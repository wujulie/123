#擷取人臉、灰階化

```python
import dlib
import cv2
from pylab import *
import os
import time

path = '/content/drive/MyDrive/基本情緒辨識教學/data/'
x = []
count = 0

for root, dirs, files in sorted(os.walk(path, topdown=False)):
    dirs.sort()
    for file in files:
        print(path + file)
        if (file == '0.jpg'):
            pass
        else:
            filepath = root + '/' + file
            detector = dlib.get_frontal_face_detector()
            img = cv2.imread(filepath)
            face = detector(img, 1)

            # 偵測人臉
            # face_rects = detector(img, 0)

            for i, d in enumerate(face):
                x1 = d.left()
                y1 = d.top()
                x2 = d.right()
                y2 = d.bottom()
                img2 = img[y1:y2, x1:x2]

            try:
                img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
                resized = cv2.resize(img2, (224, 224), cv2.INTER_LANCZOS4)

                count = count + 1

            except:
            
                pass                      
            cv2.imwrite("/content/drive/MyDrive/基本情緒辨識教學/result/0/" + file, resized)
```

```python
def batch_rename(path): #檔案重新命名
    count = 0
    for fname in os.listdir(path):
        new_fname = str(count)
        print(os.path.join(path, fname))
        os.rename(os.path.join(path, fname), os.path.join(path, new_fname))
        count = count + 1  
```
